```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Input the CSV file into python

# Compare churn count with respect to gender.

```python
plt.figure(figsize=(8,8))
sns.countplot(Telco["gender"],hue = Telco["Churn"])
plt.title("Churn count with respect to gender",fontweight="bold")
plt.grid(linestyle="--")
```

# Find out how many female senior citizens there in the dataset

```python
SCF=Telco.loc[(Telco["gender"]=="Female") & (Telco["SeniorCitizen"]==1)]
print("The Female Senior Citizens in Data set is ",SCF["gender"].count())
```

```python
plt.figure(figsize=(5,8))
sns.countplot(SCF["gender"])
plt.title("The Female Senior Citizens count in Data set is 568",fontweight="bold")
plt.yticks(np.arange(0,650,50))
plt.xlabel("Senior Citizen")
```

# Compare 'tenure' with 'Total Charges'

```python
plt.figure(figsize=(8,8))
plt.scatter(Telco["tenure"],Telco["TotalCharges"],s=10,marker="*")
plt.xlabel("Tenure")
plt.ylabel("Total Charges")
plt.title("Relation between tenure vs Total charges",fontweight="bold")
plt.yticks(np.arange(0,9000,1000))
```

# Find out which contract is preferred by the senior citizen.

```python
SC=Telco.loc[(Telco["SeniorCitizen"]==1)]
SC["Contract"].value_counts()
```

```python
plt.figure(figsize=(5,8))
sns.countplot(SC["SeniorCitizen"],hue=SC["Contract"])
plt.xticks(visible = False)
plt.title("Contract split of Senior Citizen",fontweight="bold")
plt.grid(linestyle="--")
```

```python
sns.boxplot(x="PaymentMethod",y="tenure",data=Telco)
plt.title("Box plot of Payment methods vs Tenure",fontweight="bold")
plt.xticks(rotation=10)
```

```python
#Box plot shows the distribution of the payment methods against the tenure period 
#Bank Transfer and Credit card is preferred when the Tenure is on higher side
#Electronic check and Mailed Check is preferred when the Tenure is on lower side
```
