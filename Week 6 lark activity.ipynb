# Question 2- Visualisation of the Data

```python
#Piechart showing Region wise population 
```

```python
regData = Countries.groupby("Region").sum()["Population in thousands (2017)"]
plt.figure(figsize = (10, 15))
plt.pie(regData, labels = regData.index, autopct = "%1.1f%%")
plt.title('Regionwise population',fontweight='bold',fontsize=15)
plt.show()
```

```python
#Bar Graph showing top 20 countries based on GDP
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='country',y='GDP per capita (current US$)',data=Countries.sort_values('GDP per capita (current US$)',ascending=False).head(20), palette='Set1')
plt.title('Top 20 countries based on GDP per capita (current US$)',fontweight='bold',fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('GDP per capita (current US$)',fontsize=15)
plt.xticks(rotation=90)
plt.grid(linestyle="--")
plt.show()
```

```python
#Bar Graph showing top 20 countries based on Agricultural production index
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='country',y='Agricultural production index ',data=Countries.sort_values('Agricultural production index ',ascending=False).head(20), palette='Set2')
plt.title('Top 20 countries based on Agricultural production index',fontweight='bold',fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Agricultural production index',fontsize=15)
plt.xticks(rotation=90)
plt.grid(linestyle="--")
plt.show()
```

```python
#Bar Graph showing top 20 countries based on Population 
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='country',y='Population in thousands (2017)',data=Countries.sort_values('Population in thousands (2017)',ascending=False).head(20),palette='Set3')
plt.title('Top 20 countries based on Population in thousands (2017)',fontweight='bold',fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Population in thousands (2017)',fontsize=15)
plt.xticks(rotation=90)
plt.grid(linestyle="--")
plt.show()
```

```python
#Bar Graph showing top 20 countries based on International trade: Imports (million US$)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='country',y='International trade: Imports (million US$)',data=Countries.sort_values('International trade: Imports (million US$)',ascending=False).head(20))
plt.title('Top 20 countries based on International trade:Imports (million US$)',fontweight='bold',fontsize=15)
plt.xlabel('Country',fontsize=15)
plt.ylabel('International trade: Imports (million US$)',fontsize=15)
plt.xticks(rotation=90)
plt.grid(linestyle="--")
plt.show()
```

```python
#Bar Graph showing count of countries in each region
```

```python
region = Countries['Region'].value_counts()
plt.figure(figsize=(10,8))
sns.barplot(x=region.index,y=region.values)
plt.title('Number of Countries by Region',fontweight='bold',fontsize=15)
plt.xlabel('Regions',fontsize=15)
plt.ylabel('Counts',fontsize=15)
plt.xticks(rotation=90)
plt.yticks(np.arange(0,22,2))
plt.grid(linestyle="--")
plt.show()
```

# Question 4- Top Region with Average GDP

```python
Top_GDPMAX=Countries.groupby("Region")[["GDP per capita (current US$)"]].mean()
Top_GDPMAX.sort_values(by="GDP per capita (current US$)",ascending=False)
```

```python
Top_GDPMAX.nlargest(1,"GDP per capita (current US$)")
```

# Question 4-Top Region with Average International trade: Imports (million US$)

```python
Top_ImpMAX=Countries.groupby("Region")[["International trade: Imports (million US$)"]].mean()
Top_ImpMAX.sort_values(by="International trade: Imports (million US$)",ascending=False)
```

```python
Top_ImpMAX.nlargest(1,"International trade: Imports (million US$)")
```

# Question 5-Region with most consistent GDP

```python
GP=Countries.groupby("Region")[["GDP per capita (current US$)"]].std()
GP.sort_values(by="GDP per capita (current US$)")
```

```python
GP.nsmallest(1,"GDP per capita (current US$)")
```
